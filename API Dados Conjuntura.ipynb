```python
#CAMBIO

# dt.round(freq='d') aredonda para dia sem o horario, assim pode ser para h = hora, min = min s = segundo

meses_trimestres = {
    'January':'1º Trimestre',
    'February': '1º Trimestre',
    'March': '1º Trimestre',
    'April': '2º Trimestre',
    'May': '2º Trimestre',
    'June': '2º Trimestre',
    'July': '3º Trimestre',
    'August': '3º Trimestre',
    'September': '3º Trimestre',
    'October': '4º Trimestre',
    'November': '4º Trimestre',
    'December': '4º Trimestre'    
}


link_BC_cambio = "https://olinda.bcb.gov.br/olinda/servico/PTAX/versao/v1/odata/CotacaoDolarPeriodo(dataInicial=@dataInicial,dataFinalCotacao=@dataFinalCotacao)?@dataInicial='01-01-1996'&@dataFinalCotacao='12-31-2022'&$top=10000&$format=json"
requisicao = requests.get(link_BC_cambio)
cambio = requisicao.json()

#pprint.pprint(cambio)

df_cambio = pd.DataFrame(cambio ['value'])

df_cambio['dataHoraCotacao'] = pd.to_datetime(df_cambio['dataHoraCotacao'], format = '%Y-%m-%d %H:%M:%S.%f')

df_cambio['Período'] = df_cambio['dataHoraCotacao'].dt.round(freq='d')

df_cambio['Ano'] = df_cambio['dataHoraCotacao'].dt.year
df_cambio['Mês'] = df_cambio['dataHoraCotacao'].dt.month_name().replace(nomes_meses)
df_cambio ['Trimestre'] = df_cambio['dataHoraCotacao'].dt.month_name().replace(meses_trimestres)
df_cambio = df_cambio.groupby(['Ano', 'Trimestre']).mean().reset_index()



df_cambio
```
